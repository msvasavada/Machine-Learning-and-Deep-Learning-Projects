# Dataset

A study shows that 0.01N NaOH is added to drug molecules to check its impact on PH

# Import Libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Load the Dataset and split into Dependent and independent variables

```
from google.colab import files
uploaded = files.upload()
#import io
#df2 = pd.read_csv(io.BytesIO(uploaded['Filename.csv']))
import io
# making data frame from csv file
#data = pd.read_csv("nba.csv", index_col ="Name")
dataset=pd.read_csv(io.BytesIO(uploaded['pH.csv']))
```

```
dataset
```

```
X = dataset.iloc[:, 2:-1].values
y = dataset.iloc[:, -1].values
print(X)
print(y)
```

# Training the Linear Regression model on the whole dataset

```
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

# Training the Polynomial Regression model on the whole dataset

```
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```
poly_reg7=PolynomialFeatures(degree=7)
X_poly7=poly_reg7.fit_transform(X)
lin_reg_7=LinearRegression()
lin_reg_7.fit(X_poly7,y)
```

# Visualising the Linear Regression results

```
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Drops of NaOH')
plt.ylabel('pH')
plt.show()
```

# Visualising the Polynomial Regression results

```
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Drops of NaOH')
plt.ylabel('pH')
plt.show()
```

```
#Plot all regressions
plt.scatter(X,y,color='red')
plt.plot(X, lin_reg.predict(X), color = 'cyan')
plt.plot(X,lin_reg_2.predict(poly_reg.fit_transform(X)),color='blue')
plt.plot(X,lin_reg_7.predict(poly_reg7.fit_transform(X)),color='green')
plt.title('Truth Or Bluff (Polynomial Linear Regression)')
plt.xlabel('Drops of NaOH')
plt.ylabel('pH')
plt.show()
```

# Visualising the Polynomial Regression results (for higher resolution and smoother curve)

```
X_grid = np.arange(min(X), max(X), 0.1)
#It creates an instance of ndarray with evenly spaced values and returns the reference to it.
X_grid = X_grid.reshape((len(X_grid), 1))
print(X_grid)
```

```
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Drops of NaOH')
plt.ylabel('pH')
plt.show()
```

# Predicting a new result with Linear Regression

```
y_poly4=lin_reg_2.predict(poly_reg.fit_transform([[25]]))
y_poly4
```

```
y_poly7=lin_reg_7.predict(poly_reg7.fit_transform([[25]]))
y_poly7
```

# Evaluate the models

```
from sklearn import metrics
from sklearn.metrics import mean_squared_error, r2_score
#print('Root Mean Squared Error (RMSE):', np.sqrt(metrics.mean_squared_error(y, lin_reg_2.predict(poly_reg.fit_transform([[X]]))))
print('R-Squared Value (r2) for Linear Regression :', r2_score(y, lin_reg.predict(X)))
print('R-Squared Value (r2) for Polynomial Regression (degree 4):', r2_score(y, lin_reg_2.predict(poly_reg.fit_transform(X))))
print('R-Squared Value (r2) for Polynomial Regression (degree 7):', r2_score(y, lin_reg_7.predict(poly_reg7.fit_transform(X))))
```
